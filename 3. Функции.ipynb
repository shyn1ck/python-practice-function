### 3. Функция с условным оператором

```python
price1 = 100 # цена первого продукта
num1 = 3 # его количество

price2 = 1000 # цена второго продукта
num2 = 2 # его количество

price3 = 40 # цена третьего продукта
num3 = 10 # его количество

# находим общую стоимость первой позиции
total1 = num1 * price1
if total1 > 1000: # если она превышает 1000,
    total1 -= total1 * 0.1 # то уменьшаем стоимость на 10 %
print(total1)

# находим общую стоимость второй позиции
total2 = num2 * price2
if total2 > 1000: # если она превышает 1000,
    total2 -= total2 * 0.1 # то уменьшаем стоимость на 10 %
print(total2)

# находим общую стоимость третьей позиции
total3 = num3 * price3
if total3 > 1000: # если она превышает 1000,
    total3 -= total3 * 0.1 # то уменьшаем стоимость на 10 %
print(total3)
```

```python
# объявляем функцию calculate_total_price с двумя параметрами: price и count
def calculate_total_price(price: int, count: int) -> float: # заголовок функции: ключевое слово def, имя функции и формальные параметры в скобках
    '''
    Функция предназначена для определения общей стоимости товара, принимающая следующие аргументы:

    - price - цена товара
    - count - количество товара
    '''
    total = price * count # в теле функции считаем стоимость позиции в корзине
    if total > 1000: # сравниваем стоимость с 1000 в условной конструкции
        total -= total * 0.1 # уменьшаем стоимость на 5%, если она больше 1000
    return total # возвращаем значение стоимости
#####################################################################################################################################

total1 = calculate_total_price(100, 3)
total2 = calculate_total_price(1000, 2)
total3 = calculate_total_price(40, 10)

print(f'Общая стоимость первой позиции {total1}')
print(f'Общая стоимость второй позиции {total2}')
print(f'Общая стоимость третьей позиции {total3}')
```

### 4. Функция с циклом

```python
def filter_by_year(data, year):
    '''
    Функция предназначена для фильтрации фильмов по годами и принимает аргументы:

    - data - данные в формате список внутри списка

    '''
    empty_list=[]
    for i in data:
        if i[2]>year:
            empty_list.append(i)
    return empty_list


# название, страна, год, жанр, продолжительность (мин.)
movies_table = [
    ['Побег из Шоушенка', 'США', 1994, 'драма', 142],
    ['Крёстный отец', 'США', 1972, 'драма', 175],
    ['Тёмный рыцарь', 'США', 2008, 'фантастика, боевик, триллер', 152],
    ['Список Шиндлера', 'США', 1993, 'драма', 195],
    ['Криминальное чтиво', 'США', 1994, 'триллер, комедия, криминал', 154],
    ['Хороший, плохой, злой', 'Италия', 1966, 'вестерн', 178],
    ['Бойцовский клуб', 'США', 1999, 'триллер, драма, криминал', 139],
    ['Харакири', 'Япония', 1962, 'драма, боевик, история', 133],
    ['Сталкер', 'СССР', 1979, 'фантастика, драма, детектив', 163],
    ['Иди и смотри', 'СССР', 1985, 'драма, военный', 136],
    ['Один дома', 'США', 1990, 'комедия, семейный', 103]
]

# выводим результат на экран
display(filter_by_year(movies_table, 1990))
```

### 5. Область видимости функции. Правило LEGB

Область видимости функции в Python определяет, где переменная может быть использована или изменена. В Python существует несколько уровней области видимости, и они следуют правилу LEGB (Local, Enclosing, Global, Built-in):

**Local** (Локальная область видимости): Переменные, объявленные внутри функции, являются локальными и видны только в пределах этой функции.

```python
def example_function():
    local_variable = "Local"
    print(local_variable)

example_function()
# print(local_variable)  # Ошибка, local_variable не определена вне функции
```

```python
x = 5
# x = 10
def some_func():
    global x
    x = 10

    # print(text) 
```

**Enclosing** (Объемлющая область видимости): Вложенные функции могут видеть переменные из внешних функций, в которых они объявлены.

```python
def outer_function():
    enclosing_variable = "Enclosing"

    def inner_function():
        print(enclosing_variable)  # Мы видим переменную из внешней функции

    inner_function()

outer_function()
```

**Global** (Глобальная область видимости): Переменные, объявленные за пределами всех функций, считаются глобальными и доступны в любой части программы.

```python
global_variable = "Global"

def example_function():
    global global_variable
    print(global_variable)  # Мы видим глобальную переменную

example_function()
```
